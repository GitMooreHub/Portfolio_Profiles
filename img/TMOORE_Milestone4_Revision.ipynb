## Box Plots

Box plots should show me some general distributions of data as well as representation of the data according to the density and where the densities of data points lie.

```python
# a function for the box plots - to be grouped by the group argument.
def grouped_boxplot(x,group,data):
  # First we are going to sort the index, this will serve the purpose of making sure
  # every plot shows up the same.
  sorted_index = df.groupby(group)[x].median().sort_values(ascending=True).index
  return sns.boxplot(y=group, x=x, data=data, orient = "h",order=sorted_index)

# drops columns we want for the Y axis
X = df.drop(columns=['BarDensity','BarDensityA'])

# creates the subplots for the plots
nsubplots = X.shape[1] # one subplot for each predictor
# smart way to build the exact amount of cols
ncols = round(nsubplots**0.5)
# smart way to build the exact number of rows needed
nrows = (nsubplots // ncols) + (nsubplots % ncols)
# plots each individually in the location below
position = range(1, nsubplots + 1)

# plot figures for each
fig = plt.figure(figsize=(15,15))
for i in range(nsubplots):
  ax = fig.add_subplot(nrows,ncols,position[i])
  ax = grouped_boxplot(X.columns[i],'BarDensityA',df)
  ax.set(ylabel = "Density Bins")
fig.tight_layout()
plt.show
```

##PCA

Next is PCA. This will show which variables have the most influence over the variance of the data. From here I chose my prime variables to model.

```python
from sklearn.preprocessing import StandardScaler

# standardization and PCA declaration.
scaler = StandardScaler()
pca = PCA()

# fit the data to normalization and PCA
X_scaled = scaler.fit_transform(X)
pca.fit(X_scaled)

# our loaded data into the DF to use for PCA
loadings = pd.DataFrame(pca.components_,columns=X.columns)
# explained variance ratio column
loadings['explained_var'] = pca.explained_variance_ratio_
loadings['explained_var_cumsum'] = loadings['explained_var'].cumsum()
print(loadings)
```

```python
# Now to find the top 5 absolute values per the first 12 rows (takes care of 90% of data)

# its noted in this we must take the abs in order to account for negatives
loadings.columns.values[np.argsort(abs(-loadings.values), axis=1,)[:12, -5:]]
# output will be in ascending order

# From here we can take the top 3 of each and use them
```

We see that the first principal component explains nearly 21% of the data. The first 2 show 34%. This isn't terrible, but isn't good either. It'd be better if it explained nearly 50%+.

We get nearly 70% of variance explained after the 7th principal component, and 
above 90% after the 12th. Perhaps narrowing it down to 12 variables would be best.

Starting off we will consider the top 3 values:

Our first principal component is made up of: LotNumber, DateStamp, Ni

#Model Revision

After closely examining the EDA it was concluded that some variables must go while others seem to have more impact on results than the others. For the most part however, it seems that everything pulls nearly equal weight in the final BarDensity variance.

## KNN

Firstly, we start with the KNN Model. This is because this possibly has the best chance providing details as to whether we need some intense tuning or not and where it could be tuned.

```python
# We start by checking for null values at this point.
reducedDf.isnull().sum()
```

```python
# Next, we will need some additional imports.
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
# create separate table and drop nulls (if found above)
df2 = reducedDf.dropna(axis=0)

# new list of highly interested columns
colsOfInt = ['LotNumber', 'DateStamp', 'Ni', 'Cr', 'Co', 'Mo', 'Ti', 'Ca', 'Zn', 'K', 'P', 'Bottlenumber', 'PanelNumber', 'Cu', 'Al', 'La', 'BulkDensity', 'TapDensity']

# setting x and y values
y=df2.BarDensityA
X=df2.drop(columns=[col for col in df2 if col not in colsOfInt])

# training for model
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# I pre-processed in a pipeline to avoid any leakage from the validation or test sets
# essentially what the code below does is searchest for the best values for K 
# my referencing values between 1 and 50 to determine which K gives the best score

# making pipeline
pipe = make_pipeline(
    # scales data for equal fitting
    StandardScaler(),
    # PCA extraction
    PCA(),
    # KNN model
    KNeighborsClassifier()
)

# settings up testing parameters
param_grid = dict(
    # weights
    kneighborsclassifier__weights=['uniform','distance'],
    # K values
    kneighborsclassifier__n_neighbors=np.arange(1,51)
)

# Searches using pipeline and parameters
grid_search = GridSearchCV(pipe,param_grid = param_grid)

# prints the search to make sure it worked
print(grid_search)
```

```python
# no we can fit them for the model
grid_search.fit(X_train,y_train)

# Saves the results
cv_results = pd.DataFrame({
    'CV mean accuracy' : grid_search.cv_results_['mean_test_score'],
    'weights' : [params['kneighborsclassifier__weights'] for params in grid_search.cv_results_['params']],
    'n neighbors' : [params['kneighborsclassifier__n_neighbors'] for params in grid_search.cv_results_['params']]
})
```

```python
# not to create the lineplot of the fitted model and print out
sns.lineplot(data=cv_results,x='n neighbors',y='CV mean accuracy',hue='weights')
plt.show()

print('The model that yielded the highest mean cross-validated accuracy of {0} used {1} and {2} weighting'.format(
    grid_search.best_score_,
    grid_search.best_params_['kneighborsclassifier__n_neighbors'],
    grid_search.best_params_['kneighborsclassifier__weights']
))

# gets best answer
getEst = grid_search.best_estimator_

#Now we will test a prediction using the Kneigherborsclassifier
y_pred = getEst.predict(X_test)

# prints our classification report for the model
print(classification_report(y_test,y_pred))
```

We achieved a cross-valicated accuracy of slightly above 80% using the model. our classification report doesn't look terrible either except the recall on the level 3 bar density value. This is likely do to the very limited amount of support for it. This could possibly be combated by increasing the testing size and reducing the training size, but possibly at the cost of the overall accuracy. Below is a confusion matrix of the model above, to see somewhat where our model is drifting.

```python
print(pd.DataFrame(
    confusion_matrix(y_test,y_pred),
    index = getEst.classes_,
    columns=getEst.classes_
))
```

From the confusion matrix we see that about 33% of the class 1 densities are being confused for class 2, and about 10% of class 2 are being confused for class 1. So there is a slight confusion in the model between that.

More interesting however, is that nearly half of class 3 is being confused with class 2 and even Class 1! More so than the other classes. This could be due to the bias in the data and the limited amount of information available for class 3.

From the classification report it is evident that we are in need of some major tuning as there is hardly any accuracy to the model at all. Below is a search for the best parameters and model pairing, hopefully it turns out better.

```python
# lets look for the best parameters from 3 different models
param_grid = [
   {'kernel': ['poly'],'C': [0.1,1,10,100], 'degree':[2,3,4]}, #higher degrees won't run
   {'kernel': ['rbf'],'C': [0.1,1,10,100], 'gamma': [0.001,0.01,0.1], 'class_weight':['balanced',None]},
   {'kernel': ['sigmoid'],'C':[0.1,1,10,100]}
 ]

grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)

grid.fit(X_train,y_train)
```

It looks like the best are of the original rbf model, with a gamma correction of 0.1 and a C value of only 1, while the weight chosen is none. Below is the retested model with the new parameters. Our newest model scored a huge increase in accuracy, up to nearly 83%! Below is a cross validation of this scoring.

```python
# Retesting for results
grid_predictions = grid.predict(X_test)

# print classification report
print(classification_report(y_test, grid_predictions))
```

With the improved hyperparameters it is clear to see that the model performs much better than before, although not perfectly. It is still far from a great mdoel but it is legues better than the standard un-tuned model, and may actually be able to classify the densities.
